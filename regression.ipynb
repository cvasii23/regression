### Simularea Traficului de Internet

In tema pe care am ales-o, voi simula traficul de internet folosind diferite formule,variabile si procese matematice cum ar fi variabila Poisson, procesul Pareto si forumele matematice ce le deducem din acestea.
Din codul pe care o sa il vedeti mai jos, rezulta doua fisiere text. Un fisier in care sunt salvate numarul "burst-urilor" active(sesactive.txt), iar celalalt in care avem salvate numarul de pachete trimise(pachete.txt).
Poate va ganditi ce este acela un "burst" sau la ce face referire in cazul nostru. Am sa va explic foarte simplu si la subiect: atunci cand navigam pe internet, sunt momente cand poate ne plictisim si dorim sa ascultam o piesa pe YouTube, spre exemplu, in acel moment traficul "zvacneste" si nu mai este constant. Acesta este un "burst".

Mai jos o sa atasez intreg codul ce a fost scris in Python:

```python
import math
import random
import numpy as np

global af,bt,T
T=1000.0
af=1.6
bt=float
M=7.0
L_pachet=1024
r=20000
t_pachet=L_pachet*8.0/r


def pareto():
	if (af>1):
		bt=(M*(af-1))/af
	y=random.random()
	x=bt/(1-y)**(1/af)
	return x


def proc_poiss(lim,l):
	ta=1/l
	t=0.0
	mom=[]
	while (t<lim):
		u=random.random()
		t=t-ta*math.log(u)
		mom.append(t)
	return mom
mom_init=proc_poiss(10.0, 5.0)

def pasi(lim,pas):
	mom=[]
	i=0
	while (i*pas)<=lim:
		mom.append(i*pas)
		i+=1
	return mom


tau=pasi(T,2.0)
B=[0]*len(tau)
mom_init=proc_poiss(T,5.0)

nr_init=np.random.poisson(np.floor(5.0*((M*(af-1))/af)))
C=[]

for n in range(nr_init):
	u=random.random()
	bt=(M*(af-1))/af
	if (u<1-1/af):
		x=bt*(1-(u+1/af)*af)	
	else:
		x=bt*(af*(1-u))**(1/(1-af))
		for i in range(len(tau)):
			if tau[i]<=x:
				B[i]=B[i]+1
	if int(x/t_pachet>0):
		C.append(int(x/t_pachet))

for m in mom_init:
	d=pareto()
	for i in range(len(tau)):
		if m<=tau[i]<=m+d:
			B[i]=B[i]+1
	C.append(int(d/t_pachet))

nr_ses=nr_init+len(mom_init)

file=open("sesactive.txt", "w")
for element in B:
	file.write(str(element))
	file.write(" ")
file.close()

file=open("pachete.txt","w")
for element in C:
	file.write(str(element))
	file.write(" ")
file.close()
```
